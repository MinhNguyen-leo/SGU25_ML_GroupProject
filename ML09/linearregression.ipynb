# 1. Linear Regression với ví dụ 1 biến:
$$
f(x) = (x + 5)^2
$$

### Import thư viện

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Vẽ đồ thị ước lượng min

```python
x_vals = np.arange(-10, 0.001, 0.1)
y_vals = (x_vals + 5)**2
plt.plot(x_vals, y_vals)
plt.show()
```

### Đạo hàm hàm f(x)
$$
f(x)' = \frac{df}{dx} = x(x + 5)
$$

### Cập nhật Gradient DescentL
$$
x_i = x_{i-1} - p\frac{df}{dx} = x_{i-1} - p*2(x_{i-1} + 5)
$$

### Code minh họa:

```python
x = 3
lr = 0.1
precision = 1e-4
max_iters = 1000

for i in range(max_iters):
    f_prev = (x + 5)**2
    grad = 2*(x + 5)
    x = x - lr*grad
    f_cur = (x + 5)**2
    print(f"Step {i}: x={x:.6f} --> f(x)={f_cur:.6f}")
    if abs(f_cur - f_prev) < precision:
        break

print(f"Local minimum occurs at x={x:.6f}, f(x)={f_cur:.6f}")
```
